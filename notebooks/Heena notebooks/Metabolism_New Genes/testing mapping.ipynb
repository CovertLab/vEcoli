```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd


# -----------------------------
# CONFIGURATION
# -----------------------------
REACTION_FILE = "/Users/abigailalemayeho/vEcoli/reconstruction/ecoli/flat/metabolic_reactions.tsv"
TARGET_REACTION = "ABC-32-RXN"              # <- change to reaction you want to trace

# -----------------------------
# LOAD AND PARSE DATA
# -----------------------------
def safe_json_parse(s):
    if isinstance(s, str) and s.strip().startswith("{"):
        try:
            return json.loads(s)
        except json.JSONDecodeError:
            return {}
    else:
        return {}

# Load file
df = pd.read_csv(REACTION_FILE, sep='\t', header=None,
                 names=["reaction_id", "stoichiometry", "direction", "misc", "val"])

# Parse stoichiometry safely
df["stoichiometry"] = df["stoichiometry"].apply(safe_json_parse)

# -----------------------------
# BUILD BIPARTITE GRAPH
# -----------------------------
G = nx.DiGraph()

for _, row in df.iterrows():
    rxn = row["reaction_id"]
    direction = row["direction"]
    stoich = row["stoichiometry"]

    for met, coeff in stoich.items():
        if coeff is None:
            continue  # Skip invalid coefficients
        if direction == "L2R":
            if coeff < 0:
                G.add_edge(met, rxn)
            elif coeff > 0:
                G.add_edge(rxn, met)

        elif direction == "R2L":
            if coeff > 0:
                G.add_edge(met, rxn)
            elif coeff < 0:
                G.add_edge(rxn, met)

        elif direction == "BOTH":
            if coeff < 0:
                G.add_edge(met, rxn)
                G.add_edge(rxn, met)
            elif coeff > 0:
                G.add_edge(rxn, met)
                G.add_edge(met, rxn)

# -----------------------------
# TRACE PATHWAYS FROM A REACTION
# -----------------------------
def trace_reaction_network(G, start_rxn, depth=2):
    visited = set()
    queue = [(start_rxn, 0)]
    subgraph_nodes = set()

    while queue:
        node, d = queue.pop(0)
        if node in visited or d > depth:
            continue
        visited.add(node)
        subgraph_nodes.add(node)
        neighbors = list(G.successors(node)) + list(G.predecessors(node))
        for n in neighbors:
            queue.append((n, d+1))

    return G.subgraph(subgraph_nodes)

# -----------------------------
# VISUALIZE SUBGRAPH
# -----------------------------
subG = trace_reaction_network(G, TARGET_REACTION, depth=2)

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(subG, seed=42)
node_colors = ["skyblue" if "RXN" in n else "orange" for n in subG.nodes()]
nx.draw(subG, pos, with_labels=True, node_size=800, node_color=node_colors,
        font_size=8, arrowsize=15)
plt.title(f"Local network around {TARGET_REACTION}")
plt.show()
```

```python
import pandas as pd
import networkx as nx
import json
import matplotlib.pyplot as plt
from collections import deque

# -----------------------------
# CONFIG
# -----------------------------
REACTION_FILE = "/Users/abigailalemayeho/vEcoli/reconstruction/ecoli/flat/metabolic_reactions.tsv"
START_REACTION = "ABC-32-RXN"

# -----------------------------
# HELPERS
# -----------------------------
def is_extracellular(metabolite):
    return "EXTRACELLULAR" in metabolite or metabolite.endswith("[CCO-EXTRACELLULAR]")

def clean_stoich(cell):
    try:
        return json.loads(cell) if isinstance(cell, str) and cell.strip() else {}
    except json.JSONDecodeError:
        return {}

# -----------------------------
# LOAD DATA
# -----------------------------
df = pd.read_csv(REACTION_FILE, sep='\t', header=None,
                 names=["reaction_id", "stoichiometry", "direction", "misc", "val"])
df["stoichiometry"] = df["stoichiometry"].apply(clean_stoich)

# -----------------------------
# BUILD GRAPH
# -----------------------------
G = nx.DiGraph()

for _, row in df.iterrows():
    rxn = row["reaction_id"]
    stoich = row["stoichiometry"]
    direction = row["direction"]

    for met, coeff in stoich.items():
        if coeff is None:
            continue

        if direction == "L2R":
            if coeff < 0:
                G.add_edge(met, rxn)
            elif coeff > 0:
                G.add_edge(rxn, met)

        elif direction == "R2L":
            if coeff > 0:
                G.add_edge(met, rxn)
            elif coeff < 0:
                G.add_edge(rxn, met)

        elif direction == "BOTH":
            if coeff != 0:
                G.add_edge(met, rxn)
                G.add_edge(rxn, met)

# -----------------------------
# TRAVERSAL FROM REACTION
# -----------------------------
def bfs_trace(G, start_rxn):
    visited = set()
    queue = deque([start_rxn])
    traced_edges = []

    while queue:
        node = queue.popleft()
        if node in visited:
            continue
        visited.add(node)

        for neighbor in G.successors(node):
            if is_extracellular(neighbor):
                continue  # Skip beyond extracellular nodes
            traced_edges.append((node, neighbor))
            queue.append(neighbor)

        for neighbor in G.predecessors(node):
            if is_extracellular(neighbor):
                continue
            traced_edges.append((neighbor, node))
            queue.append(neighbor)

    return traced_edges

edges = bfs_trace(G, START_REACTION)

# -----------------------------
# VISUALIZATION
# -----------------------------
def draw_subgraph(edges):
    subG = nx.DiGraph()
    subG.add_edges_from(edges)

    pos = nx.spring_layout(subG, k=0.5, iterations=100)
    node_colors = [
        'lightgreen' if 'RXN' in n else ('orange' if is_extracellular(n) else 'lightblue')
        for n in subG.nodes()
    ]

    plt.figure(figsize=(12, 8))
    nx.draw(subG, pos, with_labels=True, node_color=node_colors, node_size=500, arrows=True)
    plt.title(f"Reaction Trace from {START_REACTION}")
    plt.show()

draw_subgraph(edges)
```
