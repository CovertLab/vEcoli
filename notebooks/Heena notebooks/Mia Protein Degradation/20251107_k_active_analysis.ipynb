```python
# import tsv
file_path = 'notebooks/Heena notebooks/Mia Protein Degradation/Clim4.tsv'
df_tau = pd.DataFrame(pd.read_csv(file_path, sep='\t'))
df_tau.head()
```

# Plot 1: Scatterplot for comparing linear and ratio method for selected methods

```python
# make plotly scatterplot
import plotly.express as px
import plotly.graph_objects as go

lo = float(np.minimum(df_tau['sHL_candidate_value'].min(), df_tau['cHL_candidate_value'].min()))
hi = float(np.maximum(df_tau['sHL_candidate_value'].max(), df_tau['cHL_candidate_value'].max()))

fig = go.Figure()
fig.add_scatter(
    x=df_tau['sHL_candidate_value'], y=df_tau['cHL_candidate_value'],
    mode="markers", name="points",
    marker=dict(size=6, opacity=0.7),
    hovertemplate=f"{'sHL_candidate_value'}: %{{x:.3g}}<br>{'cHL_candidate_value'}: %{{y:.3g}}<extra></extra>"
)
# y = x identity line
fig.add_scatter(
    x=[lo, hi], y=[lo, hi],
    mode="lines", name="y = x",
    line=dict(dash="dash"),
    hoverinfo="skip"
)

fig.update_layout(
    title=f"Standard linear vs constant ratio method comparison",
    xaxis_title=f"Standard linear method",
    yaxis_title=f"Constant ratio method",
    xaxis=dict(type="log"),
    yaxis=dict(type="log"),
    template="plotly_white",
    margin=dict(l=60, r=20, t=60, b=60),
    showlegend=False
)

# fig.update_layout(
#     plot_bgcolor='rgba(0, 0, 0, 0)',  # Transparent plot area background
#     paper_bgcolor='rgba(0, 0, 0, 0)'
# )

fig.show()
# fig.write_image("notebooks/Heena notebooks/Mia Protein Degradation/out/selected_HL_method_comparison.png", scale=5, height=400, width=600)
```

# Plot 2: Barplot counting the number of methods used

```python
from collections import Counter
method_variety = Counter(df_tau.selected_HL)
fig = px.histogram(df_tau, x='selected_HL')
fig.update_xaxes(categoryorder="total descending")
fig.update_layout(template="plotly_white",
                  height=600, width=900,)
fig.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Transparent plot area background
    paper_bgcolor='rgba(0, 0, 0, 0)'
)

fig.show()
fig.write_image("notebooks/Heena notebooks/Mia Protein Degradation/out/selected_HL_method_count.png", scale=5, height=600, width=900)
```
