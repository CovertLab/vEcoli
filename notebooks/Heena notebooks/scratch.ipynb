```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
K_D = 1
V_max = 4
alpha = 2

x_arr = np.array([0, 1, 2, 4, 6, 8, 12, 16])
y_actual = (K_D/V_max)*x_arr + alpha/V_max
# y_arr1 = [yi*(1+np.random.rand()) for yi in y_arr]
# y_arr2 = [yi*(1+np.random.rand()) for yi in y_arr]
# y_arr3 = [yi*(1+np.random.rand()) for yi in y_arr]

err = 0.5
y_arr = np.array([np.random.uniform(y-err, y+err, 3) for y in y_actual])
```

```python
y_arr_copy = y_arr.copy()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_arr,y_arr[:,0], c='k', clip_on=False)
ax.scatter(x_arr,y_arr[:,1], c='k', clip_on=False)
ax.scatter(x_arr,y_arr[:,2], c='k', clip_on=False)
ax.plot(x_arr, y_actual, color='blue', label = r'$\alpha = 2$')
plt.ylim([0, 6])
plt.xlim([0, 16.5])
plt.xlabel(r'$\frac{1}{[S]} \left(\frac{1}{mM}\right)$', fontsize=15)
plt.ylabel(r'$\frac{1}{v} \left(\frac{1}{\mu M/min}\right)$', fontsize=15)
plt.legend()
plt.show()
```

```python
import itertools
line_style = itertools.cycle(('-', '--', '-.', ':'))
alpha_arr = np.array([0.5, 2, 6])

fig, ax = plt.subplots()
for a in alpha_arr:
    y_actual = (K_D/V_max)*x_arr*a + 1/V_max
    ax.plot(x_arr, y_actual, label = r'$\alpha = $' + str(a), linestyle = next(line_style))
plt.ylim([0, 6])
plt.xlim([0, 16.5])
plt.xlabel(r'$\frac{1}{[S]} \left(\frac{1}{mM}\right)$', fontsize=15)
plt.ylabel(r'$\frac{1}{v} \left(\frac{1}{\mu M/min}\right)$', fontsize=15)
plt.legend()
plt.show()
```

```python
K_D = 2
V_max = 8
alpha_arr = np.array([1, 2, 4, 8])

x_arr = np.arange(0,16,0.25)
line_style = itertools.cycle(('-', '--', '-.', ':'))
fig, ax = plt.subplots()
for a in alpha_arr:
    v_arr = (V_max * x_arr)/(K_D + x_arr*(a))
    ax.plot(x_arr, v_arr, label = r'$\alpha = $' + str(a), linestyle = next(line_style))
plt.ylim([0, 8])
plt.xlim([0, 16.5])
plt.ylabel(r'$v \left(\frac{mM}{s}\right)$', fontsize=15)
plt.xlabel(r'$[S] \left(mM\right)$', fontsize=15)
plt.legend()
plt.show()
```
