```python
df_m = df.melt(id_vars=["Media", "Condition", "Biological", "Technical"], value_name = "Concentration", variable_name = "Metabolite")

df_m = df_m.with_columns(pl.col("Concentration").str.replace_all(",", ".").str.replace_all("E", "e"))

df_m = df_m.with_columns(
    pl.col("Concentration")
        .cast(pl.Utf8)
        .str.replace_all(r"\s+", "")       # remove spaces
        .str.replace_all(",", ".")         # replace comma with period
        .str.replace_all("‐", "-")         # replace Unicode dash with regular hyphen!
        .str.strip_chars('"')              # remove quotes
        .cast(pl.Float64, strict=False)    # convert to numeric
        .alias("Concentration")
)

df_m
```

```python
df_other = pl.read_csv("park_bennett_concentrations.tsv", separator="\t", comment_prefix='#')

df_other
#old ones
```

```python
df_avg = (
    df_m.filter(pl.col("Media") == "Mineral")
    .group_by("Metabolite")
    .agg(pl.col("Concentration").mean().alias("Avg_Concentration"))
    .sort("Metabolite")
)

df_avg
```

```python
joined_df = df_other.join(df_avg, on="Metabolite", how="inner")

joined_df #this is just the ones that match
```

```python

mapping_dict = {"2‐/3‐PG": "2-PG", "G3P": "GLYCEROL-3P", "AKG" : "2-KETOGLUTARATE", "ALA" : "L-ALPHA-ALANINE", "ASP" : "L-ASPARTATE", "CGMP" : "GMP", "DADP" : "ADP", "DCDP" : "CDP", "DGDP" : "GDP", "DTDP" : "TDP", "DTTP" : "TTP", "F1,6BP": "FRUCTOSE-16-DIPHOSPHATE" , "F6P" : "FRUCTOSE-6P" , "G6P": "GLC-6-P", "GL3P" : "GLYCEROL-3P", "PEP" : "PHOSPHO-ENOL-PYRUVATE", "R5P" : "RIBOSE-5P", "S7P" : "SEDOHEPTULOSE-7P", "6PG": "6-PHOSPHOGLUCONATE DEHYDROGENASE","DTMP":"DTMP KINASE",
"F1P":"ACYL-COA THIOESTERASE", "DCMP": "DCMP PHOSPHOHYDROLASE", "GAL1P": "GALACTOSE-1-PHOSPHATE URIDYLYLTRANSFERASE", "GLU": "PHOSPHOENOLPYRUVATE CARBOXYLASE", "ICIT": "ISOCITRIC ACID" }

df_avg1 = df_avg.with_columns(pl.col("Metabolite").replace(mapping_dict))
```

```python
set(df_avg1.select("Metabolite").to_series()) - set(df_other.select("Metabolite").to_series())

```

```python
df_joined = (
    df_avg1.join(df_other, on="Metabolite", how="full")
           .with_columns(pl.coalesce([pl.col("Metabolite"), pl.col("Metabolite_right")]).alias("Metabolite"))
           .drop("Metabolite_right")
           .sort("Metabolite")
           .rename({"Avg_Concentration": "Thorfinnsdottir Concentration units.mol/units.L"})
)
df_joined
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_long = (
    df_joined.to_pandas()
    .melt(id_vars=["Metabolite"], var_name="Study", value_name="Concentration")
    .dropna(subset=["Concentration"])
)

# shorten legend labels (edit as needed)
short = {
    "Thorfinnsdottir Concentration": "Thorfinnsdottir",
    "Bennett Concentration (units: mmol/units·L)": "Bennett",
    "Park Concentration (units: μmol/units·L)": "Park",
    # ...add others exactly as they appear...
}
df_long["Study"] = df_long["Study"].replace(short)

# keep top 20 by max concentration across studies
keep = (
    df_long.groupby("Metabolite")["Concentration"]
    .max()
    .nlargest(20)
    .index
)
plot_df = df_long[df_long["Metabolite"].isin(keep)]

#order metabolites by overall mean
order = (
    plot_df.groupby("Metabolite")["Concentration"]
    .mean()
    .sort_values(ascending=False)
    .index
)

g = sns.catplot(
    data=plot_df,
    y="Metabolite", x="Concentration",
    hue="Study",
    kind="bar",
    order=order,
    height=8, aspect=1.3
)

# Move legend to the right outside the plot
plt.tight_layout()
g.set_xlabels("Concentration")
g.set_ylabels("")
plt.tight_layout()
plt.show()

```

```python
melt_df = df_joined.melt(id_vars=["Metabolite"], variable_name="Study", value_name="Concentration").drop_nulls(subset=["Concentration"]).with_columns(pl.col("Concentration").cast(pl.Float64).log())

sns.set(context="talk")
sns.catplot(y="Concentration", x="Metabolite", hue="Study", kind="point", data=melt_df.to_pandas(),height=16, aspect=1, order = order, linestyle="none", marker="_", markersize=25)

order = (
    plot_df.groupby("Metabolite")["Concentration"]
    .mean()
    .sort_values(ascending=False)
    .index
)

plt.xticks(rotation=45, ha='right')
```

```python
bennett = pl.from_pandas(plot_df).filter(pl.col("Study") == "Bennett Concentration (units.mol/units.L)")
park = pl.from_pandas(plot_df).filter(pl.col("Study") == "Park Concentration (units.mol/units.L)")
```

```python
comb = bennett.join(park, on="Metabolite", how="inner")

comb_x = comb.select(pl.col("Concentration")).to_series()
comb_y = comb.select(pl.col("Concentration_right")).to_series()
```

```python
plt.plot(comb_x, comb_y, "o")

# set log
plt.xscale("log")
plt.yscale("log")

# set fixed aspect
plt.gca().set_aspect('equal', adjustable='box')

# plot x=y
plt.plot([comb_x.min(), comb_x.max()], [comb_x.min(), comb_x.max()], 'r--')

# make axis labels
plt.xlabel("Bennett Concentration (mol/L)")
plt.ylabel("Park Concentration (mol/L)")
```
