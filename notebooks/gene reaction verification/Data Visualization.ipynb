```python
import numpy as np
import math
```

```python
reactions = fluxesWithCaption.keys()
```

```python
zeroFluxReactions = []
```

```python
nonZeroFlux = {}
```

```python
def zeroFlux(key, reactions):
    fluxes = fluxesWithCaption[key]
    for num in fluxes:
        if not num == 0:
            return False
    return True
```

```python
for key in reactions:
    if zeroFlux(key, reactions):
        zeroFluxReactions.append(key)
    else:
        res = [abs(ele) for ele in fluxesWithCaption[key]]
        nonZeroFlux[key] = res
```

```python
def Sort_Tuple(tup):
    tup.sort(key = lambda x: x[1])
    return tup

def createTuples(dictionary):
    result = []
    keys = dictionary.keys()
    for key in keys:
        mean = np.mean(dictionary[key])
        result.append(tuple((key, mean)))
    return result

tuples = createTuples(nonZeroFlux)
sortedReactionsTuples = Sort_Tuple(tuples)
```

```python
sortedReactions = {}

for x in range(len(sortedReactionsTuples)):
    elem = sortedReactionsTuples[len(sortedReactionsTuples) - x - 1]
    sortedReactions[elem[0]] = nonZeroFlux[elem[0]]
    
for key in sortedReactions.keys():
    new = []
    old = sortedReactions[key]
    for val in old:
        if val == 0:
            new.append(val)
        else:
            new.append(math.log(val))
    sortedReactions[key] = new
```
