```python
reactionDictionary = foundReactions
```

```python
import numpy as np
import math
```

```python
reactions = reactionDictionary.keys()
```

```python
len(reactions)
```

```python
zeroFluxReactions = []
```

```python
nonZeroFlux = {}
```

```python
def zeroFlux(key):
    fluxes = reactionDictionary[key]
    for num in fluxes:
        if not num == 0:
            return False
    return True
```

```python
for key in reactions:
    if zeroFlux(key):
        zeroFluxReactions.append(key)
    else:
        res = [abs(ele) for ele in reactionDictionary[key]]
        nonZeroFlux[key] = res
```

```python
transformedData = {}
for key in nonZeroFlux.keys():
    new = []
    old = nonZeroFlux[key]
    for val in old:
        new.append(math.log(val + 0.00000000000000001))
    transformedData[key] = new
```

```python
import statistics
def createTuples(dictionary):
    result = []
    keys = dictionary.keys()
    for key in keys:
        mean = statistics.mean(dictionary[key])
        result.append(tuple((key, mean)))
    return result

tuples = createTuples(transformedData)
tuples.sort(key = lambda x: x[1])
```

```python
sortedReactions = {}

for elem in tuples:
    key = elem[0]
    sortedReactions[key] = transformedData[key]
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(sortedReactions, orient='index')
df.index.name = 'Reactions'
df.reset_index(inplace=False)

df
```
