```python
complex_ids_truth = pd.read_csv(path_complex_ids)
protein_ids_truth = pd.read_csv(path_protein_ids)

complex_rich = pd.read_csv(path_complexes_rich, sep="\t")
# protein_rich = pd.read_csv(path_proteins_rich, sep="\t")

complex_minimal = pd.read_csv(path_complexes_minimal, sep="\t")
# protein_minimal = pd.read_csv(path_proteins_minimal, sep="\t")
```

```python
print(complex_ids_truth["Complex_id"].iloc[0])
print(type(complex_ids_truth["Complex_id"].iloc[0]))
```

# filtering for rich

```python
complex_names = complex_ids_truth['Complex_id'].tolist()
protein_names = protein_ids_truth['protein_id'].tolist()

columns_to_keep_complexes_rich = ['seed', 'generation'] + [col for col in complex_rich.columns if col in complex_names]
filtered_complexes_rich = complex_rich[columns_to_keep_complexes_rich]

columns_to_keep_protein_rich = ['seed', 'generation'] + [col for col in complex_rich.columns if col in protein_names]
filtered_proteins_rich = complex_rich[columns_to_keep_protein_rich]
```

# missing from rich

```python
complex_names = set(complex_ids_truth['Complex_id'])
complex_rich_cols = set(complex_rich.columns)

# Find complexes in complex_ids_truth that are NOT in complex_rich
excluded_from_truth = complex_names - complex_rich_cols

print(f"Excluded {len(excluded_from_truth)} complexes from complex_ids_truth:")
for col in excluded_from_truth:
    print(col)
```

```python
protein_names = set(protein_ids_truth['protein_id'])
protein_rich_col = set(complex_rich.columns)

# Find complexes in complex_ids_truth that are NOT in complex_rich
excluded_from_truth_protein = protein_names - protein_rich_col

print(f"Excluded {len(excluded_from_truth_protein)} proteins from complex_ids_truth:")
for col in excluded_from_truth_protein:
    print(col)
```

```python
complex_names = complex_ids_truth['Complex_id'].tolist()
protein_names = protein_ids_truth['protein_id'].tolist()

columns_to_keep_complexes_minimal = ['seed', 'generation'] + [col for col in complex_minimal.columns if col in complex_names]
filtered_complexes_minimal = complex_minimal[columns_to_keep_complexes_minimal]

columns_to_keep_protein_minimal = ['seed', 'generation'] + [col for col in complex_minimal.columns if col in protein_names]
filtered_proteins_minimal = complex_minimal[columns_to_keep_protein_minimal]
```
