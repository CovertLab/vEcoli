```python
complex_ids_truth = pd.read_csv(path_complex_ids)
protein_ids_truth = pd.read_csv(path_protein_ids)

complex_rich = pd.read_csv(path_complexes_rich, sep="\t")
# protein_rich = pd.read_csv(path_proteins_rich, sep="\t")

complex_minimal = pd.read_csv(path_complexes_minimal, sep="\t")
# protein_minimal = pd.read_csv(path_proteins_minimal, sep="\t")
```

```python
print(complex_ids_truth["Complex_id"].iloc[0])
print(type(complex_ids_truth["Complex_id"].iloc[0]))
```

# filtering for rich

```python
complex_names = complex_ids_truth['Complex_id'].tolist()
protein_names = protein_ids_truth['protein_id'].tolist()

existing_complexes_rich = [col for col in complex_names if col in complex_rich.columns]
missing_complexes_rich = [col for col in complex_names if col not in complex_rich.columns]
filtered_complexes_rich = complex_rich[['seed', 'generation'] + existing_complexes_rich].copy()
for col in missing_complexes_rich:
    filtered_complexes_rich[col] = 0
print(f"Rich complexes — added {len(missing_complexes_rich)} missing as 0: {missing_complexes_rich}")

existing_proteins_rich = [col for col in protein_names if col in complex_rich.columns]
missing_proteins_rich = [col for col in protein_names if col not in complex_rich.columns]
filtered_proteins_rich = complex_rich[['seed', 'generation'] + existing_proteins_rich].copy()
for col in missing_proteins_rich:
    filtered_proteins_rich[col] = 0
print(f"Rich proteins — added {len(missing_proteins_rich)} missing as 0: {missing_proteins_rich}")
```

# filtering for minimal

```python
existing_complexes_minimal = [col for col in complex_names if col in complex_minimal.columns]
missing_complexes_minimal = [col for col in complex_names if col not in complex_minimal.columns]
filtered_complexes_minimal = complex_minimal[['seed', 'generation'] + existing_complexes_minimal].copy()
for col in missing_complexes_minimal:
    filtered_complexes_minimal[col] = 0
print(f"Minimal complexes — added {len(missing_complexes_minimal)} missing as 0: {missing_complexes_minimal}")

existing_proteins_minimal = [col for col in protein_names if col in complex_minimal.columns]
missing_proteins_minimal = [col for col in protein_names if col not in complex_minimal.columns]
filtered_proteins_minimal = complex_minimal[['seed', 'generation'] + existing_proteins_minimal].copy()
for col in missing_proteins_minimal:
    filtered_proteins_minimal[col] = 0
print(f"Minimal proteins — added {len(missing_proteins_minimal)} missing as 0: {missing_proteins_minimal}")
```
